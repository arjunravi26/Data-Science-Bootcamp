```python
import pandas as pd
import numpy as np
```

```python
gender = ['Male','Female']
```

```python
data = {
    'Gender': np.random.choice(gender,size=20)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
unique = pd.unique(df['Gender'])
```

```python
size = df.shape[0]
```

```python
ohe_hot_data = np.zeros(shape=(size,len(unique)))
```

```python
ohe_hot_data
```

```python
pd.DataFrame([[1,2],[3,4]])
```

```python
dummy = pd.DataFrame(data=ohe_hot_data,columns=unique)
```

```python
dummy
```

```python
for i in range(len(df['Gender'])):
    if df['Gender'].iloc[i] == 'Female':
        dummy.loc[i,'Female'] = 1
    else:
        dummy.loc[i,'Male'] = 1
```
